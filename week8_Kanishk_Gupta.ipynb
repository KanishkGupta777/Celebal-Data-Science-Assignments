````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# bonus tool

def word_counter(text: str) -> int:
    """Count the number of words in a piece of text."""
    try:
        return len(text.split())
    except Exception:
        return 0
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Bonus tool -> Word Counter
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()

    # TODO: Implement routing logic

    # Example structure:
    # if "calculate" in query_lower:
    #     ...
    # elif "keywords" in query_lower:
    #     ...
    # else:
    #     ...

    print(f"received query: {query}") #for verifying what the agent decide to do

    try:
        if "calculate" in query_lower:
            print("move to Calculator Tool")
            expression = query_lower.replace("calculate", "").strip()
            answer = calculator(expression)
            if answer == "Error in calculation":
                return {"type": "error", "result": answer}
            return {"type": "calculation", "result": answer}

        elif "keywords" in query_lower:
            print("move to Keyword Tool")
            keywords = extract_keywords(query)
            return {"type": "keywords", "result": keywords}

        elif "count words" in query_lower:
            print("move to Word Counter Tool")
            count = word_counter(query)
            return {"type": "word_count", "result": count}

        else:
            print("move to General Response")
            return {"type": "general", "result": "I can help with calculations, keyword extraction, or general questions."}

    except Exception as e:
        # catch-all so the agent never crashes on a bad query
        print(f"something went wrong: {e}")
        return {"type": "error", "result": "Something went wrong while processing the query."}
````
